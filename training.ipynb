```python
import os
import sys
import json
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
from pathlib import Path
import cv2
from scipy import signal
from scipy.fft import fft
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
from tqdm import tqdm
import time

# Add to path
sys.path.insert(0, '/content')

# Import from uploaded scripts
from rppg_system import PhysNet, Config

print("="*60)
print("rPPG Heart Rate Detection - Training")
print("="*60)


print("\nDetermining model output length...")

# Create dummy model to check output size
dummy_model = PhysNet(in_channels=3)
dummy_input = torch.randn(1, 3, Config.FRAMES_PER_WINDOW, Config.IMG_SIZE, Config.IMG_SIZE)

with torch.no_grad():
    dummy_output = dummy_model(dummy_input)

MODEL_OUTPUT_LENGTH = dummy_output.shape[1]
print(f"  Model output length: {MODEL_OUTPUT_LENGTH} samples")
print(f"  Input frames: {Config.FRAMES_PER_WINDOW}")

del dummy_model, dummy_input, dummy_output


class SimpleRPPGDataset(Dataset):
    """Dataset that matches PPG length to model output"""

    def __init__(self, video_paths, ppg_signals, hr_labels):
        self.video_paths = video_paths
        self.ppg_signals = ppg_signals
        self.hr_labels = hr_labels
        self.target_length = MODEL_OUTPUT_LENGTH

        print(f"  Target PPG length: {self.target_length} (matches model output)")

    def __len__(self):
        return len(self.video_paths)

    def __getitem__(self, idx):
        frames = self.load_video(self.video_paths[idx])
        ppg = self.process_ppg(self.ppg_signals[idx])

        return {
            'frames': torch.FloatTensor(frames),
            'ppg': torch.FloatTensor(ppg),
            'hr': torch.FloatTensor([self.hr_labels[idx]])
        }

    def process_ppg(self, ppg_signal):
        """Resample PPG to match model output length"""
        ppg = np.array(ppg_signal)

        if len(ppg) == self.target_length:
            return ppg

        # Resample using linear interpolation
        x_old = np.linspace(0, 1, len(ppg))
        x_new = np.linspace(0, 1, self.target_length)
        ppg_resampled = np.interp(x_new, x_old, ppg)

        return ppg_resampled

    def load_video(self, video_path):
        """Load video frames - Returns (C, T, H, W)"""
        cap = cv2.VideoCapture(video_path)
        frames = []

        while len(frames) < Config.FRAMES_PER_WINDOW:
            ret, frame = cap.read()
            if not ret:
                break

            frame = cv2.resize(frame, (Config.IMG_SIZE, Config.IMG_SIZE))
            frame = frame.astype(np.float32) / 255.0
            frames.append(frame)

        cap.release()

        # Pad if necessary
        while len(frames) < Config.FRAMES_PER_WINDOW:
            if frames:
                frames.append(frames[-1])
            else:
                frames.append(np.zeros((Config.IMG_SIZE, Config.IMG_SIZE, 3)))

        frames = np.array(frames[:Config.FRAMES_PER_WINDOW])
        frames = np.transpose(frames, (3, 0, 1, 2))  # (C, T, H, W)

        return frames


def estimate_hr(rppg_signal, fps=30):
    """Estimate heart rate from rPPG signal"""

    if len(rppg_signal) < 30:
        return 75.0

    # Detrend
    rppg_signal = signal.detrend(rppg_signal)

    # Apply window
    rppg_signal = rppg_signal * np.hamming(len(rppg_signal))

    # FFT
    N = len(rppg_signal)
    fft_signal = fft(rppg_signal)
    psd = np.abs(fft_signal) ** 2
    freqs = np.fft.fftfreq(N, 1.0 / fps)

    # Filter to HR range
    mask = (freqs >= 0.7) & (freqs <= 3.0)
    freqs_filtered = freqs[mask]
    psd_filtered = psd[mask]

    if len(psd_filtered) == 0:
        return 75.0

    # Find peak
    peak_idx = np.argmax(psd_filtered)
    peak_freq = freqs_filtered[peak_idx]
    hr = peak_freq * 60.0

    return hr

def calculate_mae(hr_pred, hr_true):
    return np.mean(np.abs(hr_pred - hr_true))

def calculate_pearson(hr_pred, hr_true):
    if len(hr_pred) < 2:
        return 0.0
    r, _ = pearsonr(hr_pred, hr_true)
    return r

print("\nüìÇ Loading dataset...")

with open(ANNOTATIONS_PATH, 'r') as f:
    annotations = json.load(f)

video_paths = []
ppg_signals = []
hr_labels = []

for subject_name, data in annotations.items():
    video_path = DATASET_PATH / subject_name / 'vid.avi'

    if video_path.exists():
        video_paths.append(str(video_path))
        ppg_signals.append(np.array(data['ppg']))
        hr_labels.append(data['heart_rate'])

print(f"Loaded {len(video_paths)} videos")



print("\nCreating datasets...")

dataset = SimpleRPPGDataset(video_paths, ppg_signals, hr_labels)

train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size

train_dataset, val_dataset = random_split(
    dataset,
    [train_size, val_size],
    generator=torch.Generator().manual_seed(42)
)

print(f"  Train: {train_size}, Val: {val_size}")

batch_size = 4
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=0)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=0)

print("\nCreating model...")

model = PhysNet(in_channels=3)
model = model.to(Config.DEVICE)

print(f"  Parameters: {sum(p.numel() for p in model.parameters()):,}")
print(f"  Device: {Config.DEVICE}")

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)

print("\nStarting training...")
print("="*60)

num_epochs = 20
train_losses = []
val_losses = []
best_val_loss = float('inf')

local_output = Path('/content/outputs')
local_output.mkdir(parents=True, exist_ok=True)

start_time = time.time()

for epoch in range(num_epochs):
    # Training
    model.train()
    epoch_train_loss = 0.0

    train_pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Train]")
    for batch in train_pbar:
        frames = batch['frames'].to(Config.DEVICE)
        ppg_target = batch['ppg'].to(Config.DEVICE)

        optimizer.zero_grad()
        ppg_pred = model(frames)

        loss = criterion(ppg_pred, ppg_target)
        loss.backward()
        optimizer.step()

        epoch_train_loss += loss.item()
        train_pbar.set_postfix({'loss': f'{loss.item():.4f}'})

    avg_train_loss = epoch_train_loss / len(train_loader)
    train_losses.append(avg_train_loss)

    # Validation
    model.eval()
    epoch_val_loss = 0.0

    with torch.no_grad():
        val_pbar = tqdm(val_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Val]")
        for batch in val_pbar:
            frames = batch['frames'].to(Config.DEVICE)
            ppg_target = batch['ppg'].to(Config.DEVICE)

            ppg_pred = model(frames)
            loss = criterion(ppg_pred, ppg_target)

            epoch_val_loss += loss.item()
            val_pbar.set_postfix({'loss': f'{loss.item():.4f}'})

    avg_val_loss = epoch_val_loss / len(val_loader)
    val_losses.append(avg_val_loss)

    scheduler.step(avg_val_loss)

    print(f"\nEpoch {epoch+1}/{num_epochs}:")
    print(f"  Train: {avg_train_loss:.4f} | Val: {avg_val_loss:.4f}")

    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        torch.save(model.state_dict(), local_output / 'best_model.pth')
        print(f"Best model saved!")

    print("-"*60)

training_time = time.time() - start_time
print(f"\nTraining done in {training_time/60:.1f} min")

print("\nEvaluating...")

model.load_state_dict(torch.load(local_output / 'best_model.pth'))
model.eval()

all_hr_pred = []
all_hr_true = []

with torch.no_grad():
    for batch in tqdm(val_loader, desc="Evaluating"):
        frames = batch['frames'].to(Config.DEVICE)
        hr_target = batch['hr'].cpu().numpy()

        rppg_pred = model(frames).cpu().numpy()

        for i in range(len(rppg_pred)):
            hr_est = estimate_hr(rppg_pred[i], Config.FRAME_RATE)
            all_hr_pred.append(hr_est)
            all_hr_true.append(hr_target[i][0])

all_hr_pred = np.array(all_hr_pred)
all_hr_true = np.array(all_hr_true)

mae = calculate_mae(all_hr_pred, all_hr_true)
r = calculate_pearson(all_hr_pred, all_hr_true)

print("\n" + "="*60)
print("RESULTS")
print("="*60)
print(f"MAE:       {mae:.2f} bpm")
print(f"Pearson r: {r:.3f}")
print("="*60)

torch.save(model.state_dict(), local_output / 'final_model.pth')

results = {
    'mae': float(mae),
    'pearson_r': float(r),
    'train_losses': train_losses,
    'val_losses': val_losses,
    'training_time_min': training_time / 60
}

with open(local_output / 'results.json', 'w') as f:
    json.dump(results, f, indent=2)

# Plot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(train_losses, 'b-o', label='Train', linewidth=2)
axes[0].plot(val_losses, 'r-s', label='Val', linewidth=2)
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].set_title('Training')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].scatter(all_hr_true, all_hr_pred, alpha=0.6)
axes[1].plot([40, 180], [40, 180], 'r--', linewidth=2)
axes[1].set_xlabel('True HR (bpm)')
axes[1].set_ylabel('Predicted HR (bpm)')
axes[1].set_title(f'MAE: {mae:.2f}, r: {r:.3f}')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(local_output / 'results.png', dpi=150)
plt.show()

# Copy to Drive
import shutil
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

for f in ['final_model.pth', 'best_model.pth', 'results.json', 'results.png']:
    src = local_output / f
    if src.exists():
        shutil.copy(src, OUTPUT_DIR / f)

print(f"\nSaved to: {OUTPUT_DIR}")
print("\nCOMPLETE!")
```
